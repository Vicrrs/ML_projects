```python
import numpy as np
import pandas as pd
```

```python
db = pd.read_csv("/home/vicrrs/My_Projects/ML_projects/archives/svbr.csv", delimiter=';')

# Conjunto de dados
X = db.iloc[:,:].values

print(X)
```

```python
# A classe SimpleImputer é usada para tratar valores ausentes (missing values) em um conjunto de dados.
from sklearn.impute import SimpleImputer

# Fazendo calculo para preencher dados nulos da base de dados
imputer = SimpleImputer(missing_values=np.nan, strategy='median')
#  Ajuste o imputer aos dados e transforme os dados, preenchendo os valores ausentes com a mediana
imputer = imputer.fit(X[:, 1:3])
X = imputer.transform(X[:, 1:3]).astype(str)

print(X)
```

```python
# colocando na coluna 0 o nome do canal de novo
X = np.insert(X, 0, db.iloc[:,0].values, axis=1)
print(X)
```
